```python
import pandas as pd
import numpy as np
from bs4 import BeautifulSoup
import requests
import re
pd.set_option('display.max_columns', None)
```

```python
player_list = []
i = 0
while i < 30000:
    sofifa_url = "https://sofifa.com/?&showCol%5B%5D=pi&showCol%5B%5D=ae&showCol%5B%5D=hi&showCol%5B%5D=wi&showCol%5B%5D=pf&showCol%5B%5D=oa&showCol%5B%5D=pt&showCol%5B%5D=bo&showCol%5B%5D=bp&showCol%5B%5D=gu&showCol%5B%5D=vl&showCol%5B%5D=wg&showCol%5B%5D=rc&showCol%5B%5D=ta&showCol%5B%5D=cr&showCol%5B%5D=fi&showCol%5B%5D=he&showCol%5B%5D=sh&showCol%5B%5D=vo&showCol%5B%5D=ts&showCol%5B%5D=dr&showCol%5B%5D=cu&showCol%5B%5D=fr&showCol%5B%5D=lo&showCol%5B%5D=bl&showCol%5B%5D=to&showCol%5B%5D=ac&showCol%5B%5D=sp&showCol%5B%5D=ag&showCol%5B%5D=re&showCol%5B%5D=ba&showCol%5B%5D=tp&showCol%5B%5D=so&showCol%5B%5D=ju&showCol%5B%5D=st&showCol%5B%5D=sr&showCol%5B%5D=ln&showCol%5B%5D=te&showCol%5B%5D=ar&showCol%5B%5D=in&showCol%5B%5D=po&showCol%5B%5D=vi&showCol%5B%5D=pe&showCol%5B%5D=cm&showCol%5B%5D=td&showCol%5B%5D=ma&showCol%5B%5D=sa&showCol%5B%5D=sl&showCol%5B%5D=tg&showCol%5B%5D=gd&showCol%5B%5D=gh&showCol%5B%5D=gc&showCol%5B%5D=gp&showCol%5B%5D=gr&showCol%5B%5D=tt&showCol%5B%5D=bs&showCol%5B%5D=ir&showCol%5B%5D=pac&showCol%5B%5D=sho&showCol%5B%5D=pas&showCol%5B%5D=dri&showCol%5B%5D=def&showCol%5B%5D=phy&offset="+str(i)
    response = requests.get(sofifa_url)
    response_text = response.text
    soup_text = BeautifulSoup(response_text,"lxml")
    rows = soup_text.find_all("tr")
    for row in rows:
            cells = row.find_all('td')
            #print(cells)
            cells_to_string = str(cells)
            cells_sub = (re.sub(re.compile('<.*?>'),'',cells_to_string))
            #print(cells_sub)
            player_list.append(cells_sub)
            result = pd.DataFrame(player_list)
    i += 60
```

```python
player_information = player_information[0].str.split(',',expand=True)
player_information.drop(player_information.columns[0], axis=1, inplace=True)
player_information[1] = player_information[1].str.replace("\n ","")
player_information  
```

```python
player_information.drop(player_information.index[:1],inplace=True)
index=player_information[1].str.contains(r'2020\n', regex=True)
tmp = player_information.loc[index].shift(-1,axis=1)
tmp
```

```python
player_information_without_irregular =  player_information.drop(tmp.index)
player_information_concat = pd.concat([player_information_without_irregular,tmp], axis=0).reset_index()
```

```python
player_information_concat[1] = player_information_concat[1].str.replace("\n ","")
chars_to_remove = ["GK","RB","RCB","CB","LCB","LB","RWB","RDM","CDM","LDM","LWB","RM","RCM","CM","LCM","LM","LAM","CAM","RAM","RW","RF","CF","LF","LW","LS","ST","RS"]
for char in chars_to_remove:
    player_information_concat[1]=player_information_concat[1].str.replace(char,"")
    player_information_concat[1]=player_information_concat[1].str.lstrip()
player_information_concat[5] = player_information_concat[5].str.replace("\n\n\n\n","")
player_information_concat[5] = player_information_concat[5].str.split("\n").str[0]
player_information_concat[5] = player_information_concat[5].str.strip()
```

```python
#player_information_concat[6] = player_information_concat[6].replace(r'[A-Za-z]+', np.nan, regex=True)
player_information_concat[6] = player_information_concat[6].replace(r'Loan', np.nan, regex=True)
player_information_concat = player_information_concat.iloc[1:]

player_information_concat_not_na = player_information_concat[player_information_concat[6].notna()]

mask_sub = player_information_concat[6].isnull()
player_information_concat_sub = player_information_concat.loc[mask_sub].shift(-1, axis=1)

player_information_concat_sub.dropna(inplace=True)

player_information_concat_master = pd.concat([player_information_concat_not_na,player_information_concat_sub]).reset_index()
player_information_concat_master
```

```python
#Method to fix the irregularities
def fixing_columns(column, pattern):
    first_column_fix = player_information_concat_master.loc[player_information_concat_master[column].str.contains(pattern,regex=True)].shift(-1,axis=1)
    after_fix = pd.concat([player_information_concat_master,first_column_fix],axis=0)
    after_fix

    rows_to_fix = player_information_concat_master.loc[player_information_concat_master[column].str.contains(pattern,regex=True)].index.values
    player_information_concat_master.drop(rows_to_fix,inplace=True)
```

```python
fixing_columns('BOV', r'[A-Z]' )
fixing_columns('ID', r'[A-Z]' )
fixing_columns('Weight', r'(\b((?!(lbs))\w)+\b)')
```

```python
first_column_fix = player_information_concat_master.loc[player_information_concat_master["BOV"].str.contains(r'[A-Z]',regex=True)].shift(-1,axis=1)
after_fix = pd.concat([player_information_concat_master,first_column_fix],axis=0)
after_fix

rows_to_fix = player_information_concat_master.loc[player_information_concat_master["BOV"].str.contains(r'[A-Z]',regex=True)].index.values
player_information_concat_master.drop(rows_to_fix,inplace=True)
```

```python
first_column_fix = player_information_concat_master.loc[player_information_concat_master['ID'].str.contains(r'[A-Z]+', regex = True)].shift(-1,axis=1)
after_fix = pd.concat([player_information_concat_master,first_column_fix],axis=0)
after_fix

rows_to_fix = player_information_concat_master.loc[player_information_concat_master['ID'].str.contains(r'[A-Z]+',regex=True)].index.values
player_information_concat_master.drop(rows_to_fix,inplace=True)
```

```python
#player_information_concat_master[player_information_concat_master["Weight"].str.contains(r'(\b((?!(lbs))\w)+\b)',regex=True)]

first_column_fix = player_information_concat_master.loc[player_information_concat_master['Weight'].str.contains(r'(\b((?!(lbs))\w)+\b)', regex = True)].shift(-1,axis=1)
after_fix = pd.concat([player_information_concat_master,first_column_fix],axis=0)
after_fix

rows_to_fix = player_information_concat_master.loc[player_information_concat_master['Weight'].str.contains(r'(\b((?!(lbs))\w)+\b)',regex=True)].index.values
player_information_concat_master.drop(rows_to_fix,inplace=True)
```

```python
player_information_concat_master["Weight"] = round(player_information_concat_master["Weight"].str.split("lbs").str[0].astype(float)*0.45359237,0)
```

```python
player_information_concat_master["Height"] = player_information_concat_master["Height"].str[:-1]

def parse_height(height):
    # format: 5'11"
    height = height.split("'")
    feet = float(height[0])
    inch = float(height[1])
    return ((feet*30.48) + (inch*2.54))

player_information_concat_master["Height"]=round(player_information_concat_master["Height"].apply(lambda x:parse_height(x)),0)
```

```python
#Cleaning to Value column
player_information_concat_master["Value"] = player_information_concat_master["Value"].str.replace("€","")
player_information_concat_master["Value"] = player_information_concat_master["Value"].str.replace("M","")
player_information_concat_master.loc[player_information_concat_master["Value"].str.contains("K"),"Value"]=player_information_concat_master["Value"].str.split("K").str[0].astype(float)/1000
player_information_concat_master["Value"] = round(player_information_concat_master["Value"].astype(float),1)
player_information_concat_master = player_information_concat_master.loc[player_information_concat_master["Value"]>0]
```

```python
#Cleaning Wage column and convert to decimal
player_information_concat_master["Wage"] = player_information_concat_master["Wage"].str.replace("€","")
player_information_concat_master.loc[player_information_concat_master["Wage"].str.contains("K"),"Wage"]=player_information_concat_master["Wage"].str.split("K").str[0].astype(float)*1000
player_information_concat_master["Wage"] = round(player_information_concat_master["Wage"].astype(float)/1000000,5)
```

```python
#Cleaning Release_Clause column and convert to decimal
player_information_concat_master["Release Clause"] = player_information_concat_master["Release Clause"].str.replace("€","")
player_information_concat_master["Release Clause"] = player_information_concat_master["Release Clause"].str.replace("M","")
player_information_concat_master.loc[player_information_concat_master["Release Clause"].str.contains("K"),"Release Clause"]=player_information_concat_master["Release Clause"].str.split("K").str[0].astype(float)/1000
player_information_concat_master["Release Clause"] = round(player_information_concat_master["Release Clause"].astype(float),2)
```

```python
#Deleting some rows that composure columns is blank (Ronaldinho, Juninho, Xavi etc)
player_information_concat_master = player_information_concat_master.loc[player_information_concat_master["Composure"]!=" "]
```

```python
#Cleaning to IR column
player_information_concat_master["IR"]=player_information_concat_master["IR"].str[:-1]
```
