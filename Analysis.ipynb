```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
player_data = pd.read_csv('player_set.csv')
player_data.head()
```

```python
player_data.columns
```

```python
top_overall = player_data.sort_values('OVA',ascending=False).head(10)[['Name', 'Age', 'OVA']]
top_potential = player_data.sort_values('POT',ascending=False).head(10)[['Name', 'Age', 'POT']]
top_potential_by_age = player_data.sort_values(['Age','POT'],ascending=[True,False]).head(10)[['Name', 'Age', 'POT']]
top_potential_value = player_data.sort_values(['Value'],ascending=False).head(10)[['Name', 'Age', 'POT', 'Value']]
display(top_overall)
display(top_potential)
display(top_potential_by_age)
display(top_potential_value)
```

```python
#sns.set_theme(style="whitegrid")
ax = sns.barplot(x="Name", y="OVA", data=top_overall, errwidth=0.5)
plt.show()
```

```python
#Mean Value based on Position

pd.DataFrame(player_data.groupby("BP").Value.mean().sort_values(ascending=False))
```

```python
#Top 10 Clup (Mean Value)
pd.DataFrame(player_data.groupby("Team & Contract").Value.mean().sort_values(ascending=False).head(10))
```

```python
#Top 10 Clup (Total Value)

pd.DataFrame(player_data.groupby("Team & Contract").Value.sum().sort_values(ascending=False).head(10))
```
